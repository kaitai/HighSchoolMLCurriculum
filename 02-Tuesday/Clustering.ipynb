# Clustering (Unsupervised Learning)

Clustering, a group of methods in *unsupervised* learning, is a popular machine learning technique of grouping objects on the basis of some similarity. It enables us to find patterns in data even if we don't have existing labels or classifications for a training set.

We can list a few example as follows:
- Example 1: Clustering supermarket goods on the basis of their brand or utility. Or, clustering customers on the basis of their purchases.
- Example 2: Clustering students on the basis of their major.
- Example 3: Clustering plants into species on the basis of their sepal or petal length.

You might recognize the iris example in example three! In the iris dataset, experts examined the irises and labelled them as iris virginica, iris setosa, or iris versicolor using the knowledge they already had about irises. But what if you did not know there were three species and you wanted to learn about types of irises without these labels? You could use unsupervised clustering to see that there are three different types.

## Supervised Vs Unsupervised Learning

_Supervised_ learning is guided by labels that help us in pattern recognition for prediction tasks, e.g., classifying an insurance claim as fraudulent claim or not.
_Unsupervised_ learning, on the other hand, is not guided by labels and relies purely on pattern discovery.

## Iris Dataset clustering Example using K-means algorithm

K - means clustering clusters data points into _K_ clusters, where a datapoint is classified on the basis of their proximity to pre-defined K **center points**. 

First import the K-means clustering functions from scikit-learn, as well as importing numpy, our dataset, and matplotlib (which we will use for plotting later).

```python
from sklearn.cluster import KMeans
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# This line will make plots show up in our notebook instead of somewhere else:
%matplotlib inline
```

Then let's load our dataset:

```python

#import pandas as pd
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

Next, we create our k-means model, which we call model, and then fit it. Notice that we're setting the number of cluster to find to 3. Maybe we're doing that because we've looked at the data; maybe we're choosing 3 because we're lucky. Feel free to change it and see what happens!

```python
model = KMeans(n_clusters=3)
model.fit(X)
```

Last, predict labels based on the clustering algorithm and print them:

```python
labels=model.predict(X)
print(labels)
```

Let's look at a plot showing the datapoints with colors according to these labels. (That's what is happening when we write X[labels == 0,1]: we're picking out all the rows in X with label 1 in column 1.)

Notice that we have also plotted the centroid of each cluster. The "centroid" is a sort of multidimensional average, or mean. This type of model is called "k-means" because it involves grouping the data by proximity to the k centroids, or means, that naturally show up.

```python
#First two features - Sepal Length and Sepal Width are plotted

plt.scatter(X[labels == 0, 0], X[labels == 0, 1], s = 100, alpha=0.5, c = 'blue', edgecolor='k', label = 'Setosa')
plt.scatter(X[labels == 1, 0], X[labels == 1, 1], s = 100, alpha=0.5, c = 'red', edgecolor='k', label = 'Versicolour')
plt.scatter(X[labels == 2, 0], X[labels == 2, 1], s = 100, alpha=0.5, c = 'green', edgecolor='k', label = 'Virginica')

#Centroids
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:,1], s = 150, marker='*', c = 'cyan', label = 'Centers')
plt.ylim([1.5,5])
plt.xlim([3.5,11])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Clusters by Species')
plt.legend()
plt.show()
```

### Notes:

1. K must be specified (K is known as the hyperparameter).
2. Each cluster is associated with a center point (a.k.a. _centroids_).
3. It is a prototype or center based partitional algorithm.
4. Basic K-means algorithm:
    - Select initial values for centroids (they are often randomly selected at this stage)
    - Form **K** clusters by assigning data points to the closest centroid.
    - Recompute the Centroid (by taking the arithmetic mean) for each cluster.
    - Repeat steps 2 and 3 until the centroids stop changing.
5. Most common way of measuring the performance of a K-means algorithm is computing the Sum of Squared Errors (SSE). It is defined as $\sum_{i=1}^{K} \sum_{x \in C_i} dist^2(m_i,x)$ , where $x$ is data point in cluster $C_i$, and $m_i$ corresponds to mean of cluster $C_i$.

### Predicting New Labels

Of course, to put your new model to work you want to be able to predict labels for new data points you'd input. We can do that in Python by using the predict function attached to our model. For example, here are two new points or sets of measurements of sepals and petals:

```python
#Pass new values for features to predict new labels
new_points=[[5.7 ,4.4 ,1.5 ,0.4],
            [6.5 ,3.0 ,5.5 ,1.8]]
```

and here are the predictions for the species of these irises:

```python
new_labels=model.predict(new_points)
print(new_labels)
```

### Cross Tabulation

We create cross tabulation to draw inferences about our prediction results. This is a lot like the confusion matrix you encountered before -- there are many methods that are implemented in different libraries for Python, and crosstab is from Pandas.

```python
pd.crosstab(labels,y)
```

Here, the rows correspond to our predictions and columns correspond to the actual labels, because we put in "labels" first and "y" second in crosstab(labels,y). We notice, species corresponding to actual label 0 (Setosa, in the first column) is well clustered, while clusters with actual labels 1 (Versicolor) and 2 (Virginica) are confounded to a little extent.

### Evaluating Performance

Another methodology to measure the performance is to compute the inertia from sklearn library. Smaller inertia implies denser clusters. Inertia is defined as the distance from the centroid, i.e., SSE.

```python
print(model.inertia_)
```

### Good value of K

How do we choose the right value for $k$? We can use something called the elbow method: try many values of $k$ and look at the sum of squared errors (SSE). The more clusters you choose, the smaller the sum of squared errors, but that doesn't necessarily correspond to the best answer. In the "elbow method" we look at which value of $k$ seems to correspond with the best accuracy for the least complexity:

```python
SSE = []

for i in range(1, 10):
    model = KMeans(n_clusters = i)
    model.fit(X)
    SSE.append(model.inertia_)
    
#Find 'The elbow' or 'the knee' value of K
plt.figure()
plt.plot(range(1,10),SSE,'-ro',markevery=[2])
plt.xlim(0,10)
plt.ylim(0,700)
plt.xticks(np.arange(0, 10, 1)) 
plt.xlabel('Number of clusters')
plt.ylabel('SSE') #Inertia
plt.show()
```

According to the 'elbow method', we choose the minimum value of $k$ that gives us a reasonably low value of inertia/SSE. We try to keep the value of $k$ low as we do not want a complex model with too many clusters. 


## Hierarchical Clustering

Hierarchical Clustering is a methodology wherein we build a hierarchy of clusters. Here, the clusters at the lower end of hierarchy are contained within bigger clusters at the upper end of the hierarchy. E.g., species within genus.

```python
#<img src='https://github.com/sharm636/DaycampTA/blob/master/Iris_dendrogram.png'>
#<img src='https://github.com/sharm636/DaycampTA/blob/master/clustert.png'>
```

- We start with each object belonging to their own cluster. Clusters are formed by merging the two most similar object.
- We continue this process until all clusters are merged into one big cluster.
- This is called "agglomerative" clustering, where we start from bottom and go upwards. It is the most common type of clustering. There is another method called "divisive" clustering as well, which works the other way around -- splitting clusters apart rather than joining them together.

```python
#Data from Example 1 of https://stackabuse.com/hierarchical-clustering-with-python-and-scikit-learn/
X = np.array([[5,3],  
    [10,15],
    [15,12],
    [24,10],
    [30,30],
    [85,70],
    [71,80],
    [60,78],
    [70,55],
    [80,91],])
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='average')  
cluster.fit_predict(X) 
print(cluster.labels_) 
```

```python
plt.scatter(X[:,0],X[:,1], c=cluster.labels_, cmap='rainbow')  
```

### Dendrogram

```python
labels = range(1, 11)  
plt.figure(figsize=(10, 7))  
plt.subplots_adjust(bottom=0.1)  
plt.scatter(X[:,0],X[:,1], label='True Position')

for label, x, y in zip(labels, X[:, 0], X[:, 1]):  
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show()  
```

Label the points as above before drawing the dendrogram. We will be using agglomerative approach, so we will be starting with individual points and make clusters on our way up.

```python
from scipy.cluster.hierarchy import dendrogram, linkage  
linked = linkage(X, 'average')

labelList = range(1, 11)

plt.figure(figsize=(10, 7))  
dendrogram(linked,  
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()  
```

A dendrogram is a graphical representation of a tree diagram that shows how all clusters are tied together. Notice, we start at the bottom and make our way up by merging two points/clusters together. Depending on where you decide to 'cut' the tree, you will get different number of clusters. Cutting at y=100 gives you one cluster. Cutting at y=40 gives you two clusters. Cutting at y=1 gives you 10 clusters. Which do you think is best?
